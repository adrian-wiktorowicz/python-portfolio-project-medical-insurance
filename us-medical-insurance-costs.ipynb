# U.S. Medical Insurance Costs

# Finding out the average age of the patients in the dataset.

```python
import csv
import matplotlib.pyplot as plt
```

```python
insurance_list = []
count = 0
with open('insurance.csv', newline='') as insurance:
    reader = csv.DictReader(insurance)



    for row in reader:
        count += 1
        insurance_list.append(row)

    
```

```python
# print(insurance_list[0].get('age'))

total_age = 0

for i in range(len(insurance_list)):
    total_age += int(insurance_list[i].get('age'))

average_age = total_age / count

print(f'Average age of patients in data set is: {average_age}')

```

```python
northeast_count = 0
southwest_count = 0

for i in range(len(insurance_list)):
    region = insurance_list[i].get('region')
    if region == 'southwest':
        southwest_count += 1
    elif region == 'northeast':
        northeast_count += 1

print(northeast_count)
print(southwest_count)

labels = 'Southwest', 'Northeast'
sizes = [northeast_count, southwest_count]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```

```python
smoker_count = 0
no_smoker_count = 0

smoker_total_insurance = 0
no_smoker_total_insurance = 0

for i in range(len(insurance_list)):
    smoker = insurance_list[i].get('smoker')
    if smoker == 'yes':
        insurance_cost = float(insurance_list[i].get('charges'))
        smoker_total_insurance += insurance_cost
        smoker_count += 1
    elif smoker == 'no':
        insurance_cost = float(insurance_list[i].get('charges'))
        no_smoker_total_insurance += insurance_cost
        no_smoker_count += 1

labels = 'Smoker', 'No smoker'
sizes = [smoker_count, no_smoker_count]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.show()

avg_smoker_insurance = smoker_total_insurance / smoker_count
print(avg_smoker_insurance)
avg_no_smoker_insurance = no_smoker_total_insurance / no_smoker_count
print(avg_no_smoker_insurance)

data = {'Smoker': smoker_count, 'No smoker': no_smoker_count}
keys = ['Smoker insurance', 'No smoker insurance']
values = [avg_smoker_insurance, avg_no_smoker_insurance]

plt.bar(keys, values, color='maroon', width=0.4)
plt.ylabel("Avg cost of insurances in their own category")
plt.show()
```
